```python
import spacy
nlp = spacy.load("en_core_sci_sm")
nlp_eng = spacy.load('en_core_web_sm')
from scispacy.abbreviation import AbbreviationDetector
abbr_pip = AbbreviationDetector(nlp)
nlp.add_pipe(abbr_pip)
```

```python

doc = nlp(text)
```
