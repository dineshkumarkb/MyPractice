```python
#Extracting named entity through web scraping
from bs4 import BeautifulSoup
import requests
import re
```

```python
def url_str(url):
    res = requests.get(url)
    html = res.text
    soup = BeautifulSoup(html,'html.parser')
    for script in soup(["script","style","aside"]):
        script.decompose()
    return ' '.join(re.split(r'[\n\t]+',soup.get_text()))
        
    
#clean_txt = url_str(r'https://www.nytimes.com/2018/08/13/us/politics/peter-strzok-fired-fbi.html?hp&action=click&pgtype=Homepage&clickSource=story-heading&module=first-column-region&region=top-news&WT.nav=top-news')
clean_txt = url_str(r'https://en.wikipedia.org/wiki/Rajinikanth')
```

```python
news = nlp(clean_txt)
for X in news.ents:
    print((X.text,X.label_))
```

```python
sentences = [x for x in news.sents]
```

```python
displacy.render(nlp(str(sentences[0:30])), jupyter=True, style='ent')
```
